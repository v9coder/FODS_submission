# DATA CLEANING


```python
import pandas as pd
df = pd.read_csv('/content/AirQuality.csv', sep = ';')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
print(df.columns)
```

```python
numeric_columns = ['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)',
                   'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)',
                   'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['CO(GT)'])
plt.title('Box Plot of CO(GT) Before Removing Outliers')
plt.ylabel('CO(GT) Levels')
plt.show()
```

# DATA Transformation

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

```python
df.to_csv('modified_air_quality_data.csv', index=False)
```
