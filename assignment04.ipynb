```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the normal distribution
mean = 100
std_dev = 4
size = 100000

# Generate the normal distribution data
data = np.random.normal(mean, std_dev, size)

# Plotting the histogram and KDE
sns.histplot(data, bins=30, kde=True, stat='density', color='blue', alpha=0.6)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution (Mean = 100, SD = 4)')
plt.axvline(mean, color='red', linewidth=1)  # Mean line
plt.axvline(mean + std_dev, color='green', linewidth=1)  # 1 SD line
plt.axvline(mean - std_dev, color='green', linewidth=1)  # -1 SD line
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 100
std_dev = 4
size = 100000

# Generate the normal distribution data
data = np.random.normal(mean, std_dev, size)

# Standardize the data
standardized_data = (data - mean) / std_dev

# Plotting the histogram and KDE for the standard normal distribution
sns.histplot(standardized_data, bins=30, kde=True, stat='density', color='blue', alpha=0.6)
plt.xlabel('Z-Value')
plt.ylabel('Density')
plt.title('Standard Normal Distribution (Mean = 0, SD = 1)')

# Add lines for the mean (0) and 1 standard deviation
plt.axvline(0, color='red', linewidth=1)  # Mean line
plt.axvline(1, color='green', linewidth=1)  # 1 SD line
plt.axvline(-1, color='green', linewidth=1)  # -1 SD line

plt.show()
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean = 100
std_dev = 4
size = 100000

data = np.random.normal(mean, std_dev, size)

# QQ-Plot for the normal distribution
fig = sm.qqplot(data, line='s')
plt.title('QQ-Plot for Normal Distribution (Mean = 100, SD = 4)')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid()
plt.show()
```
